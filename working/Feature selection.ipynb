# Feature Selection

```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
```

```python
df_crime_socioecon = pd.read_csv('data\Chicago_Merged.csv').drop(columns = 'Unnamed: 0')
df_census  = pd.read_csv('data\Census.csv')
df = pd.merge(df_crime_socioecon, df_census, on ='Community Area', how = 'outer')
df.dropna(inplace = True)
df.head(3)
```

```python
df.columns.values
```

```python
ord_enc = OrdinalEncoder()
arrest_ord_enc = ord_enc.fit_transform(df['Arrest'].values.reshape(-1, 1))
domestic_ord_enc = ord_enc.fit_transform(df['Domestic'].values.reshape(-1, 1))
```

```python
df.columns.values
features = df[[ 'ID','Arrest', 'Domestic', 'Community Areas', 
        'Latitude','Longitude', 'PERCENT OF HOUSING CROWDED',
       'PERCENT HOUSEHOLDS BELOW POVERTY', 'PERCENT AGED 16+ UNEMPLOYED',
       'PERCENT AGED 25+ WITHOUT HIGH SCHOOL DIPLOMA',
       'PERCENT AGED UNDER 18 OR OVER 64', 'PER CAPITA INCOME ',
       'HARDSHIP INDEX', 'Total Population',
       'Racial Demographics - Hispanic Proportion',
       'Racial Demographics - Black Proportion',
       'Racial Demographics - White Proportion',
       'Racial Demographics - Asian Proportion',
       'Racial Demographics - Other or Multiple Races Proportion']]
target = df['Primary Type']
```
